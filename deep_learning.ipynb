# üß† What is ANN?

It is the fundamental unit of deep learning. It mimics the brain's structure using neurons (nodes) arranged in layers.

It‚Äôs the foundation ‚Äî the original and simplest form of neural networks.

# üß© Structure of ANN

**Input Layer ‚Äì** Takes in features (data).

**Hidden Layers ‚Äì** Fully connected layers with neurons that do math (weights, biases, activations).

**Output Layer ‚Äì** Final prediction (class, regression value, etc.).

# üîπ Key Concepts

**Neurons and Layers-** Each neuron performs a weighted sum of inputs + bias ‚Üí passed through an activation function.

**Activation Functions-** ReLU, Sigmoid, Tanh, Softmax (used to introduce non-linearity).

**Loss Function-** Measures the difference between predicted and actual values. (e.g., MSE, Cross-Entropy)

**Backpropagation-** The core training algorithm that updates weights using gradients from the loss function.

**Optimizer-** Algorithms like SGD, Adam, etc., adjust weights to minimize loss.

# üî∑ Step-by-Step Theory to Build an ANN (Artificial Neural Network) from Scratch

---

## 1Ô∏è‚É£ Input Layer

The model receives raw input data (features).  
Example with 3 features:

\[
X = \begin{bmatrix} x_1 & x_2 & x_3 \end{bmatrix}
\]

---

## 2Ô∏è‚É£ Weights and Bias Initialization

Each neuron has:  
- **Weights**: Influence of each input  
- **Bias**: Adjusts the output  

Initialize:

\[
W = \begin{bmatrix} w_1 & w_2 & w_3 \end{bmatrix}, \quad b = b
\]

---

## 3Ô∏è‚É£ Forward Propagation

\[
z = W^T \cdot X + b = w_1x_1 + w_2x_2 + w_3x_3 + b
\]

Apply activation:

\[
a = \text{activation}(z)
\]

### Common Activation Functions:
- **ReLU**: \( \max(0, z) \)
- **Sigmoid**: \( \frac{1}{1 + e^{-z}} \)
- **Tanh**: \( \frac{e^z - e^{-z}}{e^z + e^{-z}} \)

---

## 4Ô∏è‚É£ Hidden Layers

For each hidden layer \( l \):

\[
z^{[l]} = W^{[l]} \cdot a^{[l-1]} + b^{[l]}
\]  
\[
a^{[l]} = \text{activation}(z^{[l]})
\]

---

## 5Ô∏è‚É£ Output Layer

Depends on the task:

- **Sigmoid** ‚Üí Binary classification  
- **Softmax** ‚Üí Multi-class classification  
- **Linear** ‚Üí Regression

\[
\hat{y} = \text{activation}(z^{[output]})
\]

---

## 6Ô∏è‚É£ Loss Function

Compare prediction \( \hat{y} \) with true label \( y \):

- **MSE (Regression)**:
\[
\text{Loss} = \frac{1}{n} \sum (y - \hat{y})^2
\]

- **Cross-Entropy (Classification)**:
\[
\text{Loss} = -\sum \left[ y \cdot \log(\hat{y}) + (1 - y) \cdot \log(1 - \hat{y}) \right]
\]

---

## 7Ô∏è‚É£ Backpropagation

Compute gradients:

\[
\frac{\partial \text{Loss}}{\partial W}, \quad \frac{\partial \text{Loss}}{\partial b}
\]

Use **chain rule** to propagate error backward from output to input.

---

## 8Ô∏è‚É£ Update Weights ‚Äì Gradient Descent

\[
W = W - \alpha \cdot \frac{\partial \text{Loss}}{\partial W}
\]  
\[
b = b - \alpha \cdot \frac{\partial \text{Loss}}{\partial b}
\]

Where \( \alpha \) is the **learning rate**.

---

## üîÅ Training Loop

Repeat steps **3‚Äì8** over multiple epochs until the model's performance stabilizes.

**Simple Regression**

```python
import logging
logging.basicConfig(level=logging.INFO, format='%(message)s', force=True)

import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score

# -------------------- Load Data --------------------
def load_data(url: str):
    try:
        data = pd.read_csv(url)
        logging.info('‚úÖ Data loaded successfully!')
        return data
    except Exception as e:
        logging.error("‚ùå Error loading data", exc_info=True)
        return None

# -------------------- Scale Features --------------------
def process_data(df):
    df.dropna(inplace=True)
    x = df[['Open']]
    y = df['Next_Close']

    x_scaler = MinMaxScaler()
    y_scaler = MinMaxScaler()

    X_scaled = x_scaler.fit_transform(x)
    y_scaled = y_scaler.fit_transform(y.values.reshape(-1, 1))

    logging.info(f"üìä Scaled X shape: {X_scaled.shape}")
    logging.info(f"üéØ Scaled y shape: {y_scaled.shape}")
    return X_scaled, y_scaled, x_scaler, y_scaler

# -------------------- Build Model --------------------
def build_model(input_dim):
    model = Sequential([
        Dense(64, input_dim=input_dim, activation='relu'),
        Dense(64, activation='relu'),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    logging.info("‚úÖ Model built and compiled.")
    return model

# -------------------- Train Model --------------------
def train_model(model, X_train, y_train, X_val, y_val):
    early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=20,
        batch_size=32,
        callbacks=[early_stop],
        verbose=1
    )
    logging.info("‚úÖ Training completed.")
    return history

# -------------------- Make Predictions --------------------
def make_predictions(model, X_val, y_scaler):
    pred_scaled = model.predict(X_val)
    pred_unscaled = y_scaler.inverse_transform(pred_scaled)
    return pred_unscaled

# -------------------- Custom Prediction --------------------
def predict_custom_price(model, x_scaler, y_scaler, open_price):
    open_scaled = x_scaler.transform(np.array([[open_price]]))
    pred_scaled = model.predict(open_scaled)
    pred_price = y_scaler.inverse_transform(pred_scaled)[0][0]
    logging.info(f"üîÆ Prediction for Open={open_price}: Next_Close={pred_price:.2f}")
    print(f"\nüîÆ Custom Prediction:\nOpen: {open_price} ‚û°Ô∏è Predicted Next Close: {pred_price:.2f}")
    return pred_price

# -------------------- Evaluate --------------------
def evaluate_model(y_true_scaled, y_pred_scaled, y_scaler):
    y_true = y_scaler.inverse_transform(y_true_scaled.reshape(-1, 1))
    y_pred = y_scaler.inverse_transform(y_pred_scaled.reshape(-1, 1))

    rmse = mean_squared_error(y_true, y_pred)**0.5
    r2 = r2_score(y_true, y_pred)

    logging.info(f"üìä RMSE: {rmse:.4f}")
    logging.info(f"üìà R¬≤ Score: {r2:.4f}")

    print(f"\nüìä RMSE: {rmse:.4f}")
    print(f"üìà R¬≤ Score: {r2:.4f}")

# -------------------- Plot --------------------
def plot_predictions(y_val_scaled, y_pred_unscaled, y_scaler):
    y_true = y_scaler.inverse_transform(y_val_scaled.reshape(-1, 1))
    y_pred = y_pred_unscaled

    plt.figure(figsize=(10, 5))
    plt.plot(y_true, label='Actual')
    plt.plot(y_pred, label='Predicted')
    plt.title('üìà Actual vs Predicted - Stock Prices')
    plt.xlabel('Samples')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()

# -------------------- Main --------------------
def main():
    url = "https://raw.githubusercontent.com/AqueeqAzam/data-science-and-machine-learning-datasets/refs/heads/main/regression_dl.csv"
    df = load_data(url)

    if df is not None:
        X_scaled, y_scaled, x_scaler, y_scaler = process_data(df)

        X_train, X_val, y_train, y_val = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

        model = build_model(X_train.shape[1])
        train_model(model, X_train, y_train, X_val, y_val)

        loss, mae = model.evaluate(X_val, y_val)
        print(f"\n‚úÖ Final Evaluation:\nLoss: {loss:.4f}, MAE: {mae:.4f}")

        predictions = make_predictions(model, X_val, y_scaler)
        evaluate_model(y_val, predictions, y_scaler)
        plot_predictions(y_val, predictions, y_scaler)

        # Optional: Custom Prediction
        open_price_input = 1920.50
        predict_custom_price(model, x_scaler, y_scaler, open_price_input)

if __name__ == "__main__":
    main()
```

**Multi Regresson**

```python
import logging
logging.basicConfig(level=logging.INFO, format='%(message)s', force=True)

import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping

import matplotlib.pyplot as plt

# -------------------- Load Data --------------------
def load_data(url: str):
    try:
        data = pd.read_csv(url)
        logging.info('‚úÖ Data loaded successfully!')
        return data
    except Exception as e:
        logging.error("‚ùå Error loading data", exc_info=True)
        return None

# -------------------- Clean Data --------------------
def clean_data(df):
    df.dropna(inplace=True)
    X = df.drop(columns=['Date', 'Close'])
    y = df['Next_Close']

    X = X.select_dtypes(include=[np.number])
    imputer = SimpleImputer()
    X_imputed = imputer.fit_transform(X)

    logging.info(f"üîç Imputed X shape: {X_imputed.shape}")
    return X_imputed, y.values

# -------------------- Scale Features --------------------
def process_data(X, y):
    x_scaler = MinMaxScaler()
    y_scaler = MinMaxScaler()

    X_scaled = x_scaler.fit_transform(X)
    y_scaled = y_scaler.fit_transform(y.reshape(-1, 1))

    logging.info(f"üìä Scaled X shape: {X_scaled.shape}")
    logging.info(f"üéØ Scaled y shape: {y_scaled.shape}")
    return X_scaled, y_scaled, x_scaler, y_scaler

# -------------------- Build Model --------------------
def build_model(input_dim):
    model = Sequential([
        Dense(64, input_dim=input_dim, activation='relu'),
        Dense(64, activation='relu'),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    logging.info("‚úÖ Model built and compiled.")
    return model

# -------------------- Train Model --------------------
def train_model(model, X_train, y_train, X_val, y_val):
    early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=20,
        batch_size=32,
        callbacks=[early_stop],
        verbose=0
    )
    logging.info("‚úÖ Training completed.")
    return history

# -------------------- Evaluate Model --------------------
def evaluate_model(model, X_val, y_val, y_scaler):
    loss, mae = model.evaluate(X_val, y_val, verbose=0)
    pred_scaled = model.predict(X_val, verbose=0)
    pred_unscaled = y_scaler.inverse_transform(pred_scaled)
    y_true = y_scaler.inverse_transform(y_val.reshape(-1, 1))

    rmse = mean_squared_error(y_true, pred_unscaled)**0.5 # Calculate RMSE without 'squared' parameter
# r2 = r2_score(y_true, pred_unscaled)

    r2 = r2_score(y_true, pred_unscaled)

    logging.info(f"\nüìä Final Evaluation:")
    logging.info(f"   - Loss: {loss:.4f}")
    logging.info(f"   - MAE: {mae:.4f}")
    logging.info(f"   - RMSE: {rmse:.4f}")
    logging.info(f"   - R¬≤ Score: {r2:.4f}")
    return y_true, pred_unscaled

# -------------------- Plot Predictions --------------------
def plot_predictions(y_true, y_pred):
    plt.figure(figsize=(10, 5))
    plt.plot(y_true, label='Actual')
    plt.plot(y_pred, label='Predicted')
    plt.title('üìà Actual vs Predicted - Stock Prices')
    plt.xlabel('Samples')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()

# -------------------- Custom Prediction --------------------
def predict_custom(model, my_input, x_scaler, y_scaler):
    custom_scaled = x_scaler.transform([my_input])
    pred_scaled = model.predict(custom_scaled, verbose=0)
    pred_actual = y_scaler.inverse_transform(pred_scaled)
    logging.info(f"\nüí∞ Predicted Next Close Price (custom input): {pred_actual[0][0]:.2f}")
    return pred_actual[0][0]

# -------------------- Run Full Pipeline --------------------
def run_pipeline():
    url = "https://raw.githubusercontent.com/AqueeqAzam/data-science-and-machine-learning-datasets/refs/heads/main/regression_dl.csv"
    df = load_data(url)

    if df is None:
        return

    X, y = clean_data(df)
    X_scaled, y_scaled, x_scaler, y_scaler = process_data(X, y)

    X_train, X_val, y_train, y_val = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

    model = build_model(X_train.shape[1])
    train_model(model, X_train, y_train, X_val, y_val)

    y_true, y_pred = evaluate_model(model, X_val, y_val, y_scaler)
    plot_predictions(y_true, y_pred)

    # Custom input for prediction (based on feature order)
    my_input = [121.5, 126.8, 119.2, 1.4e6, 120.7]  # Example values
    predict_custom(model, my_input, x_scaler, y_scaler)

# -------------------- Main --------------------
def main():
    run_pipeline()

if __name__ == "__main__":
    main()
```

Deep Learning Algorithms

2Ô∏è‚É£ Convolutional Neural Networks (CNNs)
---
üìå Use Case: Use Image_Path for image classification.
üìå Best When:
‚úÖ The dataset consists of images, videos, or visual data.
‚úÖ The task requires object detection, image classification, or segmentation.

3Ô∏è‚É£ Recurrent Neural Networks (RNNs)
---
üìå Use Case: Use Review (text) for sentiment analysis or time series prediction.
üìå Best When:
‚úÖ The data is sequential (e.g., text, time-series, speech data).
‚úÖ You need memory across previous time steps (e.g., forecasting).

‚ö†Ô∏è Avoid When:
‚ùå The dataset is too large (RNNs struggle with long sequences; LSTMs work better).
‚ùå There are vanishing gradient problems (deep sequences require LSTMs or Transformers).

üîß Industry Applications:

Chatbots & Voice Assistants (Google Assistant, Siri, Alexa) ‚Üí Understanding natural language in conversations.
Stock Price Prediction (Finance AI) ‚Üí Forecasting future stock trends.
Machine Translation (Google Translate, DeepL) ‚Üí Converting text from one language to another.
4Ô∏è‚É£ Long Short-Term Memory Networks (LSTMs) & Gated Recurrent Units (GRUs)
---
üìå Use Case: Use Review (text) for sentiment analysis or Monthly_Sales for time-series forecasting.
üìå Best When:
‚úÖ The dataset has long-term dependencies (e.g., long sentences, time-series with past trends).
‚úÖ The problem requires remembering patterns from long sequences.

‚ö†Ô∏è Avoid When:
‚ùå You need real-time predictions (LSTMs are slower than Transformers).
‚ùå The problem doesn‚Äôt involve sequential data.

üîß Industry Applications:

Fraud Detection (Banking, PayPal, Stripe, Amazon Payments) ‚Üí Detecting fraud based on transaction history.
Financial Market Forecasting (Bloomberg, Goldman Sachs, Hedge Funds) ‚Üí Stock market trends & economic forecasting.
